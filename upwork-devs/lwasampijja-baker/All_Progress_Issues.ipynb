```
#@title
import requests
import json
import os
import numpy as np
import pandas as pd 
from pandas import json_normalize
import datetime as dt 
import datetime

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.expand_frame_repr', False)
```

#### Select the repos for preview

```
repos =['k8-test-data','s-k8-releases','k8-data-visualization','p-k8-jmeter-test-engine-v1', 's-k8-proxy-rebuild','k8-traffic-generator',
        'k8-reverse-proxy','k8-glasswall-rebuild','s-k8-buildpipeline', 'k8-performance-use-cases','k8-proxy-desktop', 's-k8-templates', 'k8-security'
        ,'k8-proxy-marketing', 'rebuild-k8s-filetypedetection']
```

#### Table Showing the Progress off All Issues

```
#@title
dfs         = []
for repo in repos:
  jtoken              = os.getenv('GITHUB_TOKEN', '')
  headers             = {'Authorization': f'token {jtoken}'}
  url                 = f'https://api.github.com/repos/filetrust/{repo}/issues'
  res                 = requests.get(url, headers=headers, params={'state': 'all'}).json()
  data                = json_normalize(res, max_level=1)
  temp_df             = pd.DataFrame(data)
  temp_df['repo']     = repo
  dfs.append(temp_df)

                
df                    = pd.concat(dfs, ignore_index=True)
df                    = df.loc[df['state'] == 'open']
df['created_at']      = pd.to_datetime(df['created_at']).dt.date
df['Aging Days']      = datetime.datetime.now().date() - df['created_at'] 
df                    = df[['repo','title','assignee.login', 'created_at', 'Aging Days']]
df = df.rename({'repo':'Project', 'title':'Issue Title', 'assignee.login':'Assignees', 'created_at':'Date Created'}, axis=1)
df
      
```
